```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

# =====================================
# LOAD REPORT
# =====================================

student_report = pd.read_csv("Student_Attendance_Report.csv")

# =====================================
# DISPLAY DATA
# =====================================

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

print("Total Students:", len(student_report))
display(student_report)

# =====================================
# SEARCH STUDENT
# =====================================

name = input("Enter Student Name: ")

result = student_report[
    student_report["Name"].str.contains(
        name,
        case=False,
        na=False
    )
]

print("\nStudent Details:")
display(result)
```

```python
# =====================================
# PIE CHART
# =====================================

status_counts = (
    student_report["Certificate_Status"]
    .value_counts()
)

plt.figure(figsize=(6,6))

plt.pie(
    status_counts.values,
    labels=status_counts.index,
    autopct="%1.1f%%"
)

plt.title("Certificate Eligibility Distribution")

plt.show()
```

```python
# =====================================
# BAR CHART
# =====================================

plt.figure(figsize=(6,4))

plt.bar(
    status_counts.index,
    status_counts.values
)

plt.xlabel("Certificate Status")
plt.ylabel("Number of Students")
plt.title("Eligible vs Not Eligible Students")

plt.show()
```

```python
# =====================================
# SCATTER PLOT
# =====================================

plt.figure(figsize=(10,6))

plt.scatter(
    student_report["Sessions_Attended"],
    student_report["Attendance_Percentage"]
)

plt.xlabel("Sessions Attended")
plt.ylabel("Attendance Percentage")
plt.title("Attendance Analysis")

plt.grid(True)

plt.show()
```

```python
# =====================================
# TOP 20 STUDENTS
# =====================================

top20 = student_report.sort_values(
    by="Attendance_Percentage",
    ascending=False
).head(20)

plt.figure(figsize=(12,6))

plt.bar(
    top20["Name"],
    top20["Attendance_Percentage"]
)

plt.xticks(rotation=90)

plt.xlabel("Student Name")
plt.ylabel("Attendance Percentage")
plt.title("Top 20 Students by Attendance")

plt.show()

print("\nAnalysis Completed Successfully!")
```
